```python
import tkinter as tk
from tkinter import messagebox

def power(x, y, p):
    res = 1
    x = x % p
    if x == 0:
        return 0

    while y > 0:
        if y & 1 == 1:
            res = (res * x) % p
        y = y >> 1
        x = (x * x) % p
    return res

def calculate_security_data():
    try:
        p = int(entry_p.get())
        alpha = int(entry_alpha.get())
        a = int(entry_a.get())
        k = int(entry_k.get())
        x = int(entry_x.get())

        beta = power(alpha, a, p)
        y1 = power(alpha, k, p)
        y2 = x * power(beta, k, p) % p
        e = p - 1 - a
        b = power(y1, e, p)
        result_b = b % p
        result_x = (y2 * b) % p

        result_label.config(text=f"Key (beta): {beta}\nY1: {y1}\nY2: {y2}\nB: {result_b}\nX: {result_x}")
    except ValueError:
        messagebox.showerror("Error", "Vui lòng nhập giá trị số hợp lệ.")

# Create main window
root = tk.Tk()
root.title("Security Data Transmission")

# Create and place labels, entries, and button
label_p = tk.Label(root, text="Nhập giá trị p:")
label_p.grid(row=0, column=0, padx=10, pady=5)
entry_p = tk.Entry(root)
entry_p.grid(row=0, column=1, padx=10, pady=5)

label_alpha = tk.Label(root, text="Nhập giá trị alpha:")
label_alpha.grid(row=1, column=0, padx=10, pady=5)
entry_alpha = tk.Entry(root)
entry_alpha.grid(row=1, column=1, padx=10, pady=5)

label_a = tk.Label(root, text="Nhập giá trị a:")
label_a.grid(row=2, column=0, padx=10, pady=5)
entry_a = tk.Entry(root)
entry_a.grid(row=2, column=1, padx=10, pady=5)

label_k = tk.Label(root, text="Nhập giá trị k:")
label_k.grid(row=3, column=0, padx=10, pady=5)
entry_k = tk.Entry(root)
entry_k.grid(row=3, column=1, padx=10, pady=5)

label_x = tk.Label(root, text="Nhập giá trị x:")
label_x.grid(row=4, column=0, padx=10, pady=5)
entry_x = tk.Entry(root)
entry_x.grid(row=4, column=1, padx=10, pady=5)

calculate_button = tk.Button(root, text="Tính toán", command=calculate_security_data)
calculate_button.grid(row=5, column=0, columnspan=2, pady=10)

result_label = tk.Label(root, text="")
result_label.grid(row=6, column=0, columnspan=2, pady=5)

# Run the Tkinter event loop
root.mainloop()
```
