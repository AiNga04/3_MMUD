```python
import tkinter as tk
from tkinter import messagebox

def calculate_inverse_and_power():
    try:
        p = int(entry_p.get())
        q = int(entry_q.get())
        e = int(entry_e.get())
        m = int(entry_m.get())

        inverse = mod_inverse(e, (q - 1) * (p - 1))
        power_result = power(m, e, p * q)

        result_label.config(text=f"Modulo Inverse: {inverse}\nPower Result: {power_result}")
    except ValueError:
        messagebox.showerror("Error", "Please enter valid integer values for p, q, e, and m.")

def mod_inverse(A, M):
    for X in range(1, M):
        if (((A % M) * (X % M)) % M == 1):
            return X
    return -1

def power(x, y, p):
    res = 1
    x = x % p
    if x == 0:
        return 0

    while y > 0:
        if y & 1 == 1:
            res = (res * x) % p
        y = y >> 1
        x = (x * x) % p
    return res

# Create main window
root = tk.Tk()
root.title("Modulo Inverse and Power Calculator")

# Create and place labels, entries, and button
label_p = tk.Label(root, text="Enter p:")
label_p.grid(row=0, column=0, padx=10, pady=5)
entry_p = tk.Entry(root)
entry_p.grid(row=0, column=1, padx=10, pady=5)

label_q = tk.Label(root, text="Enter q:")
label_q.grid(row=1, column=0, padx=10, pady=5)
entry_q = tk.Entry(root)
entry_q.grid(row=1, column=1, padx=10, pady=5)

label_e = tk.Label(root, text="Enter e:")
label_e.grid(row=2, column=0, padx=10, pady=5)
entry_e = tk.Entry(root)
entry_e.grid(row=2, column=1, padx=10, pady=5)

label_m = tk.Label(root, text="Enter m:")
label_m.grid(row=3, column=0, padx=10, pady=5)
entry_m = tk.Entry(root)
entry_m.grid(row=3, column=1, padx=10, pady=5)

calculate_button = tk.Button(root, text="Calculate", command=calculate_inverse_and_power)
calculate_button.grid(row=4, column=0, columnspan=2, pady=10)

result_label = tk.Label(root, text="")
result_label.grid(row=5, column=0, columnspan=2, pady=5)

# Run the Tkinter event loop
root.mainloop()
```
